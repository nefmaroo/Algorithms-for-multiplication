```python
"""only one experiment for each length"""
import matplotlib.pyplot as plt
import csv
x = []
y = []
y2 = []
y3 = []

with open('/Users/macbook/Desktop/dat.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(int(row[0]))
        y.append(float(row[1]))
        y2.append(float(row[2]))
        y3.append(float(row[3]))

plt.plot(x,y,label = "The Grade School Multiplication algorithm ")
plt.plot(x,y2,label = "The Divide-and-Conquer Multiplication algorithm")
plt.plot(x,y3,label = "Karatsuba Multiplication algorithm")




plt.title('The dependence of the speed of the algorithms on the value of the variable')

plt.xlabel('x - the amount of the digits of numbers')
plt.ylabel('y - time required')
plt.legend(loc='best') 
plt.show()
```
